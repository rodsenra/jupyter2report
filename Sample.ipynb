# Sample report snippet

This sample shows that we can embed comments to be rendered with a report.
Cells tagged with caption will be rendered in the report as rendered markdown.

```python
# this cell has no tag so it will not be rendered in the report
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Dummy code just to random an embedded image
# the output of this cell will be tagged with `chart` and thus rendered in the report
import numpy as np
import matplotlib.pyplot as plt

# Reproducible random data
np.random.seed(42)
x = np.arange(0, 100)
# cumulative sum creates a smoother random walk
y = np.cumsum(np.random.randn(100))

plt.figure(figsize=(10, 4))
plt.plot(x, y, color='#1f77b4', linewidth=2, label='Random walk')
plt.scatter(x, y, s=10, color='#1f77b4', alpha=0.6)
plt.title('Random Line Chart')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
